```python
# EDA_Detailed.ipynb
# ---------------------------------------------
# Расширенный анализ данных (CICIDS2017)
# ---------------------------------------------
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Чтобы графики отображались прямо в ноутбуке:
%matplotlib inline

# 1. Загрузка данных
# Предположим, что у нас есть объединённый CSV после data_loader или вы хотите взять каждую часть по отдельности.
# Здесь для примера возьмём объединённый CICIDS2017 (допустим, вы создали "full_data.csv").

data_path = "../data/raw"   # путь к сырым данным
file_list = [
    "Monday-WorkingHours.pcap_ISCX.csv",
    "Tuesday-WorkingHours.pcap_ISCX.csv",
    "Wednesday-workingHours.pcap_ISCX.csv",
    "Thursday-WorkingHours-Morning-WebAttacks.pcap_ISCX.csv",
    "Thursday-WorkingHours-Afternoon-Infilteration.pcap_ISCX.csv",
    "Friday-WorkingHours-Morning.pcap_ISCX.csv",
    "Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv",
    "Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv"
]

df_list = []
for f in file_list:
    full_path = os.path.join(data_path, f)
    if os.path.exists(full_path):
        temp = pd.read_csv(full_path, low_memory=False)
        # Удалим пробелы в названиях колонок
        temp.columns = temp.columns.str.strip()
        df_list.append(temp)
    else:
        print(f"Файл не найден: {full_path}")

if len(df_list) == 0:
    print("Нет данных для анализа.")
    import sys
    sys.exit(0)

df = pd.concat(df_list, ignore_index=True)
print(f"Считано всего {len(df_list)} файлов. Объединённый размер: {df.shape}")

# 2. Общий обзор
print("Столбцы набора данных:")
print(df.columns.tolist())

print("Первые 5 строк:")
display(df.head())

# 2.1. Количество пропусков, типы данных
print("Информация о DataFrame:")
df.info()

print("\nЧисло пропусков в каждом признаке:")
print(df.isna().sum())

# 3. Визуальный анализ распределения метки (Label)
# Предположим, что Label хранит 'BENIGN' или тип атаки
print("\nРаспределение значений в Label:")
print(df["Label"].value_counts())

# Кодируем метку в 0/1 для упрощения EDA
df["LabelEncoded"] = df["Label"].apply(lambda x: 0 if x == "BENIGN" else 1)

# 4. Статистика по числовым признакам
numeric_cols = df.select_dtypes(include=[np.number]).columns
print("\nОписание числовых признаков:")
display(df[numeric_cols].describe().T)

# 5. Пример чистки и обработки пропусков/Inf
# Предположим, что "Flow Bytes/s" бывает NaN или Inf
if "Flow Bytes/s" in df.columns:
    # Заполняем пропуски медианой
    median_val = df["Flow Bytes/s"].median(skipna=True)
    df["Flow Bytes/s"] = df["Flow Bytes/s"].fillna(median_val)
    # Заменяем Inf на NaN и удаляем
    df["Flow Bytes/s"] = df["Flow Bytes/s"].replace([np.inf, -np.inf], np.nan)
    df.dropna(subset=["Flow Bytes/s"], inplace=True)
    print("Обработали пропуски и Inf для 'Flow Bytes/s'.")

# 6. Посмотрим на некоторые гистограммы
# Сфокусируемся на нескольких важных признаках
focus_features = [
    "Max Packet Length",
    "Packet Length Mean",
    "Flow Bytes/s",   # если такой признак действительно есть
    "Flow Packets/s", # и т.д.
]

for feat in focus_features:
    if feat in df.columns:
        plt.figure(figsize=(8,4))
        benign_data = df[df["LabelEncoded"] == 0][feat].dropna()
        attack_data = df[df["LabelEncoded"] == 1][feat].dropna()
        sns.histplot(benign_data, color="green", kde=True, label="BENIGN", stat="density")
        sns.histplot(attack_data, color="red", kde=True, label="ATTACK", stat="density")
        plt.title(f"Распределение признака '{feat}' (BENIGN vs ATTACK)")
        plt.xlabel(feat)
        plt.ylabel("Плотность")
        plt.legend()
        plt.show()

# 7. Boxplot для просмотра выбросов
# Например, "Max Packet Length"
if "Max Packet Length" in df.columns:
    plt.figure(figsize=(6,4))
    sns.boxplot(data=df, x="LabelEncoded", y="Max Packet Length")
    plt.title("Boxplot: Max Packet Length (0=BENIGN,1=ATTACK)")
    plt.show()

# 8. Корреляционная матрица
# Берём часть числовых признаков (чтобы не было слишком громоздко)
some_numeric_cols = focus_features.copy()
some_numeric_cols.append("LabelEncoded")

# Проверяем, что все колонки действительно есть
some_numeric_cols = [c for c in some_numeric_cols if c in df.columns]

corr_matrix = df[some_numeric_cols].corr()
plt.figure(figsize=(6,5))
sns.heatmap(corr_matrix, annot=True, cmap="Blues", fmt=".2f")
plt.title("Корреляционная матрица (выбранные признаки)")
plt.show()

# 9. Дополнительный анализ распределения метки в разрезе SourceFile (дни недели)
# (Часто в CICIDS2017 интересно посмотреть, какие атаки на какой день приходятся)
if "SourceFile" in df.columns:
    pivot_counts = df.groupby(["SourceFile", "Label"]).size().reset_index(name="Count")
    print("\nРаспределение метки по файлам:")
    display(pivot_counts)
    # Можно построить barplot
    plt.figure(figsize=(10,4))
    sns.countplot(data=df, x="SourceFile", hue="Label")
    plt.xticks(rotation=45)
    plt.title("Количество потоков по файлам (дням) и метке")
    plt.show()

# 10. Итоговые выводы о структуре, проблемах и т.п.
# Можно сделать итоговый print или markdown-ячейку, описав, что увидели.

print("\nИтоговая информация об анализе данных:")
print("- Набор данных содержит ~{} строк.".format(df.shape[0]))
print("- Метка (Label) распределена следующим образом:")
print(df["Label"].value_counts())
print("- Мы видим, что 'Max Packet Length' и 'Flow Bytes/s' сильно скошены: ")
print("  присутствует множество значений, близких к 0, и небольшая часть высоких.")
print("- Для улучшения качества последующих моделей будем масштабировать признаки,")
print("  заполнять пропуски, и возможно удалять редкие аномальные выбросы (при необходимости).")

# Можно сохранить промежуточный датасет (после очистки) в processed-папку, если нужно:
# df.to_csv("../data/processed/full_data_cleaned.csv", index=False)
```
