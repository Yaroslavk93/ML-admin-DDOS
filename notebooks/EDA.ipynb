В этом ноутбуке мы:
1. Загрузим обработанные данные (или сырые).
2. Посмотрим распределения признаков.
3. Построим гистограммы, диаграммы размаха (Boxplot), тепловую карту корреляций.
4. Проанализируем баланс классов.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

X_train = pd.read_csv("../data/processed/X_train.csv")
y_train = pd.read_csv("../data/processed/y_train.csv")
y_train_original = pd.read_csv("../data/processed/y_train_original.csv")

X_train['Label'] = y_train
X_train['OriginalLabel'] = y_train_original

# Фильтруем DDoS
ddos_data = X_train[X_train['OriginalLabel'] == "DDoS"]
benign_data = X_train[X_train['OriginalLabel'] == "BENIGN"]

sns.histplot(benign_data['Flow Bytes/s'], color='green', label='BENIGN', kde=True)
sns.histplot(ddos_data['Flow Bytes/s'], color='red', label='DDoS', kde=True)
plt.title("Распределение Flow Bytes/s для BENIGN vs DDoS")
plt.legend()
plt.show()
```
