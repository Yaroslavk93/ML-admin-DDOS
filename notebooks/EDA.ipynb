В этом ноутбуке мы:
1. Загрузим обработанные данные.
2. Посмотрим распределения признаков.
3. Проанализируем один из признаков и смотрим, как различается его распределение для нормального трафика (BENIGN) и атак (ATTACK). Данный анализ поможет понять, можно ли использовать данный признак для разделения классов.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

X_train = pd.read_csv("../data/processed/X_train.csv")
y_train = pd.read_csv("../data/processed/y_train.csv")

# Добавляем колонку Label к X_train
X_train['Label'] = y_train['Label']

# 0 = BENIGN, 1 = ATTACK (все атаки будут вместе, без разделения по типам)
benign_data = X_train[X_train['Label'] == 0]
attack_data = X_train[X_train['Label'] == 1]

# Выбираем доступный признак, например 'Max Packet Length'
feature = 'Max Packet Length'
print("Доступные признаки в X_train:", X_train.columns.tolist())

plt.figure(figsize=(10,5))
sns.histplot(benign_data[feature], color='green', kde=True, label='BENIGN')
sns.histplot(attack_data[feature], color='red', kde=True, label='ATTACK')
plt.title(f"Распределение '{feature}' для BENIGN vs ATTACK")
plt.xlabel(feature)
plt.ylabel("Частота")
plt.legend()
plt.show()
```
