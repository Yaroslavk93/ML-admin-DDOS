```python
# DetailedSelectedFeaturesEDA.ipynb
# ------------------------------------------------------------------
# Анализ 30 отобранных признаков из CICIDS2017 (после feature selection).
# ------------------------------------------------------------------
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os

%matplotlib inline

# 1) Определяем список 30-ти признаков, которые мы уже отобрали в проекте
selected_features = [
    "Packet Length Std",
    "Avg Bwd Segment Size",
    "Max Packet Length",
    "Destination Port",
    "Average Packet Size",
    "Bwd Packet Length Max",
    "Bwd Packet Length Std",
    "Packet Length Variance",
    "Bwd Packet Length Mean",
    "Init_Win_bytes_forward",
    "Packet Length Mean",
    "Total Fwd Packets",
    "Fwd Packet Length Mean",
    "Avg Fwd Segment Size",
    "Fwd Packet Length Max",
    "Total Length of Fwd Packets",
    "Init_Win_bytes_backward",
    "Subflow Fwd Packets",
    "PSH Flag Count",
    "Subflow Fwd Bytes",
    "act_data_pkt_fwd",
    "Fwd IAT Std",
    "Bwd Packets/s",
    "Idle Mean",
    "Bwd Packet Length Min",
    "Fwd Packets/s",
    "Total Backward Packets",
    "Flow Packets/s",
    "ACK Flag Count",
    "Fwd IAT Max"
]

# 2) Считываем данные из подготовленных файлов (X_train.csv, y_train.csv)
#    Предположим, что в data/processed у нас есть обучающий набор, куда уже попали только эти 30 признаков
processed_path = "../data/processed"
X_train_path = os.path.join(processed_path, "X_train.csv")
y_train_path = os.path.join(processed_path, "y_train.csv")

X_train = pd.read_csv(X_train_path)
y_train = pd.read_csv(y_train_path)

print("X_train shape:", X_train.shape, "y_train shape:", y_train.shape)

# Убедимся, что у нас действительно есть все 30 признаков в X_train
missing_in_X = set(selected_features) - set(X_train.columns)
if len(missing_in_X) > 0:
    print("Некоторые отобранные признаки отсутствуют в X_train:", missing_in_X)
else:
    print("Все 30 отобранных признаков присутствуют в X_train.")

# Объединим для удобства с меткой (0=BENIGN, 1=ATTACK)
X_train["Label"] = y_train.values

# 3) Быстрый обзор статистики
print("\n--- Описательная статистика по 30 признакам ---")
display(X_train[selected_features].describe().T)

# 4) Проверим, есть ли пропуски или нулевые значения
print("\nЧисло пропусков в отобранных признаках:")
print(X_train[selected_features].isna().sum())

# Иногда мы можем проверить, нет ли подозрительных 'Inf' (хотя на этапе предварительной обработки их уже удаляли).
# Если бы хотели, то:
# for col in selected_features:
#     inf_count = X_train[col].isin([np.inf, -np.inf]).sum()
#     if inf_count > 0:
#         print(f"{col} содержит Inf: {inf_count} значений.")

# 5) Распределение метки (BENIGN vs ATTACK) в train-выборке
print("\nРаспределение класса Label (0=BENIGN, 1=ATTACK):")
print(X_train["Label"].value_counts())

# 6) Посмотрим гистограммы и KDE для каждого признака
#    Поскольку признаков 30, можно сделать перебор.
#    Для ясности разобьём на несколько подгрупп, чтобы графики не захламляли вывод.

def plot_distribution(df, feature, bins=50):
    """
    Для одного признака строим наложенные гистограммы (или KDE)
    для BENIGN vs ATTACK.
    """
    plt.figure(figsize=(8,4))
    sns.histplot(
        df[df["Label"] == 0][feature], color="green",
        bins=bins, label="BENIGN", kde=True, stat="density"
    )
    sns.histplot(
        df[df["Label"] == 1][feature], color="red",
        bins=bins, label="ATTACK", kde=True, stat="density"
    )
    plt.title(f"Распределение '{feature}' (BENIGN vs ATTACK)")
    plt.xlabel(feature)
    plt.ylabel("Плотность")
    plt.legend()
    plt.show()

# Разобьём 30 признаков на 2-3 части по 10, чтобы нагляднее смотреть
subset_1 = selected_features[:10]
subset_2 = selected_features[10:20]
subset_3 = selected_features[20:30]

for feat in subset_1:
    plot_distribution(X_train, feat)

for feat in subset_2:
    plot_distribution(X_train, feat)

for feat in subset_3:
    plot_distribution(X_train, feat)

# 7) Анализ выбросов (Outliers).
#    Для каждого признака можно применить правило IQR (Q1-1.5IQR, Q3+1.5IQR),
#    Посчитаем долю выбросов для BENIGN и ATTACK отдельно.

def calc_outliers_iqr(series):
    """Возвращает (кол-во выбросов, доля выбросов) по IQR-критерию."""
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    mask_out = (series < lower) | (series > upper)
    count_out = mask_out.sum()
    out_ratio = count_out / len(series)
    return count_out, out_ratio

print("\n--- Анализ выбросов (IQR) для каждого признака ---")
outliers_info = []
for feat in selected_features:
    col_benign = X_train.loc[X_train["Label"] == 0, feat]
    col_attack = X_train.loc[X_train["Label"] == 1, feat]

    count_out_benign, ratio_benign = calc_outliers_iqr(col_benign)
    count_out_attack, ratio_attack = calc_outliers_iqr(col_attack)
    outliers_info.append({
        "feature": feat,
        "count_out_benign": count_out_benign,
        "ratio_out_benign": ratio_benign,
        "count_out_attack": count_out_attack,
        "ratio_out_attack": ratio_attack
    })

outliers_df = pd.DataFrame(outliers_info)
display(outliers_df.sort_values("ratio_out_benign", ascending=False))

# 8) Boxplot для визуального обзора выбросов (возьмём 5-6 признаков, где их доля максимальна)
outliers_sorted = outliers_df.sort_values("ratio_out_benign", ascending=False).head(6)
top_outliers_feats = outliers_sorted["feature"].tolist()
print(f"\nНаибольшее число выбросов по BENIGN у признаков: {top_outliers_feats}")

for feat in top_outliers_feats:
    plt.figure(figsize=(6,4))
    sns.boxplot(data=X_train, x="Label", y=feat)
    plt.title(f"Boxplot: {feat} (Label=0 BENIGN, 1 ATTACK)")
    plt.show()

# 9) Корреляционная матрица (по всем 30 признакам).
#    Можно посмотреть на самые сильные корреляции.
corr_matrix = X_train[selected_features].corr()
plt.figure(figsize=(15,12))
sns.heatmap(corr_matrix, cmap="viridis", annot=False)
plt.title("Корреляционная матрица отобранных 30 признаков")
plt.show()

# 10) Выводы по анализу

print("\n--- Предварительные выводы: ---")
print("1) По гистограммам видно, что многие признаки (например, Max Packet Length, Packet Length Std, etc.)")
print("   имеют сильно скошенное распределение: большая часть значений сосредоточена у нуля, но встречаются достаточно высокие пики.")
print("2) Разница между BENIGN и ATTACK распределениями у ряда признаков (Avg Bwd Segment Size, Fwd Packet Length Max и т.п.)")
print("   заметна (ATTACK часто имеет пик в районе 0.0), но встречаются и пересекающиеся диапазоны.")
print("3) Анализ IQR-порогов показал, что у ряда признаков доля выбросов может достигать 10-15% (особенно Idle Mean, Bwd Packet Length Min).")
print("   Это может означать, что данные действительно содержат экстремальные значения, частично соответствующие атакам,")
print("   а частично — аномалиям в нормальном трафике.")
print("4) Корреляционная матрица показывает, что часть признаков (Packet Length Std, Packet Length Variance, Average Packet Size)")
print("   может быть сильно взаимосвязана между собой. Это стоит учитывать, чтобы не было избыточных признаков или мультиколлинеарности.")
print("5) Следующим шагом может быть дополнительная фильтрация выбросов или принятие решения оставить их, ведь в anomaly detection")
print("   выбросы как раз могут быть полезны для обучения модели VAE/ISO.")
print("6) В целом, датасет после предварительной очистки выглядит достаточно корректно: пропусков нет (или немного), Inf не встречается.")
print("   Но распределения явно не гауссовые — большая часть признаков сильно скошена, что оправдывает MinMaxScaler (а не StandardScaler).")

print("\nЗавершённый обзор 30 отобранных признаков. ")
```
